## **Data Cleaning**

### **Null Values**

```python
df.isnull().sum()
```

```python
# Filling with median values
df['Review Rating'] =  df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#Dropping promo code used column
df = df.drop('promo_code_used',axis = 1)
```

```python
df.columns
```

### **Data Visualizations**

```python
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

```

```python
# Gender Distribution
gender_count = df['gender'].value_counts()
plt.bar(gender_count.index,gender_count)
plt.title("Gender Distribution")
plt.xlabel("gender")
plt.ylabel("count")
plt.show()
```

```python
df['payment_method'].value_counts().plot(kind='bar')
plt.title('Most Used Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Performing Categories by Revenu
categories_by_sales = df.groupby('category')['purchase_amount'].agg('sum').sort_values()
plt.barh(categories_by_sales.index,categories_by_sales)
plt.xlabel("purchase amount")
plt.ylabel("categories")
plt.title('Top Performing Categories by Revenu')
plt.show()
```

```python
# Top 10 Purchased Items
Most_Purchased_Items = df['item_purchased'].value_counts().head(10)
plt.barh(Most_Purchased_Items.index,Most_Purchased_Items)
plt.title('Top 10 Purchased Items')
plt.xlabel('count')
plt.ylabel("items")
plt.show()
```

```python
# Seasonal Sales Trend
season_trends = df.groupby('season')['purchase_amount'].agg('sum')
plt.bar(season_trends.index,season_trends)
plt.xlabel("seasons")
plt.ylabel("purchase amount")
plt.title("Seasonal Sales Trend")
plt.show()
```

```python
df['location'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Number of Customers by Location')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='category', y='purchase_amount', data=df)
plt.xticks(rotation=45)
plt.title('Purchase Amount by Category')
plt.show()
```

```python

gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()
```

```python
payment_counts = df['payment_method'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Distribution')
plt.show()
```

```python
subscription_counts = df['subscription_status'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', startangle=90, colors=['blue','red'])
plt.title('Subscription Status Distribution')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Multivariate)')
plt.show()

```

```python
sns.boxplot(x='category', y='purchase_amount', hue='gender', data=df)
plt.title('Purchase Amount by Category and Gender')
plt.xlabel('Category')
plt.ylabel('Purchase Amount')
plt.xticks(rotation=45)
plt.legend(title='Gender', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

```python
sns.pairplot(df[['age', 'purchase_amount', 'review_rating', 'previous_purchases']])
plt.suptitle('Multivariate Relationships Between Variables', y=1.02)
plt.show()
```
